## 1) Problem Statement:
* Aim: The goal of the project is to develop a classification model that could help identify the type of exploitation that a victim of human trafficking would be experiencing based on various input data about a victim.  This could potentially be helpful to organizations that help in fighting this cause to identify victims and better allocate resources.  Additionally, this could also give policy makers useful information to focus legislation on where it would have the most impact on reducing the crime as well as assisting the victims.
* Hypothesis: The hypothesis is that there are associations between the type of exploitation versus input factors such as gender, age, citizenship, means of control, recruiter relationship, and country of exploitation.  With these input factors, type of exploitation can be modeled and predicted.

## 2) Potential Methods and Models:
* Since this is a classification problem, here are some potential methods/models that could be used:
    * Logistic Regression
    * K-Nearest Neighbors
    * Random Forest
    * Decision Trees
    * Naive Bayes
* The Train-Test Split method will be utilized when creating these classification models.  Cross validation will be incorporated to limit the problem of overfitting

## 3) Data - CTDC Global Dataset:
* Explanation of Data: The main data set used in this project would be from the Global Dataset from Counter-Trafficking Data Collaborative (CTDC), which serves as a central repository for critical information about human trafficking.  It publishers standard and harmonized data from various organizations using unified schema in its Global Dataset.  Current contributors include International Organization for Migration (IOM) and Polaris Project, who began the collaboration by comparing their existing data models and data classification systems and identified areas that were identical or compatible.  They then agreed on a shared lexicon and format.  
* The Global Dataset consist of information on identified and reported victims of human trafficking.  The data has                 been k-anonymized (k=11), and currently has approximately 47,000 observations.  There are 62 variables, that capture information on the social-demographics profile of victims (such as age and gender), the trafficking process (such as means of control used on victims), and the exploitation type.
* Links to Data:
    * https://www.ctdatacollaborative.org/dataset/resource/511adcb7-b1a2-4cc7-bf2f-0960d43a49cc#%7Bview-graph:%7BgraphOptions:%7Bhooks:%7BprocessOffset:%7B%7D,bindEvents:%7B%7D%7D%7D%7D,graphOptions:%7Bhooks:%7BprocessOffset:%7B%7D,bindEvents:%7B%7D%7D%7D%7D

## 5) Outstanding Questions: Here are some assumptions, risks, and caveats on this project:
* One assumption is that the sample is significant enough to be representative of the population.  Data and studies based on data.gov and UNdata above reflect the difficulty of estimating the scope of human trafficking.  Some barriers to obtain accurate estimates of the prevalence of the issue includes:
    * Data are frequently program-specific and duplicates - CTDC is the start of an effort in data collaboration between organizations and programs, but the journey of such unification is going to be challenging given the globality of the issue
    * Estimates only include women and children for sexual exploitation - There could biases in the data obtained
    * Inconsistent definition of human trafficking - More unified ways of definition have been developed over the years but often times there are still confusion and gray areas such as the comparison between trafficking and smuggling.
    * Human trafficking is an underreported crime - the datasets might only be touching the tip of the iceberg. Access to traffickers’ information are often limited. Therefore, the picture described by the data might not be completely holistic.
* Due to the sensitivity of data related to trafficked victims and survivors, the datasets tend to be semi-aggregate and lack complete granularity.  For example, the CTDC data has been k-anonymized (k=11), and therefore, certain granular data has been lost.  This caveat is shared between all datasets this project is based on and an important one to consider when proposing hypothesis and drawing conclusions.
